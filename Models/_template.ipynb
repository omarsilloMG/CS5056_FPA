```python
import pandas as pd
```

```python
data_folder       = 'data/'
database_filename = 'custom_database.csv'

df = pd.read_csv(data_folder + database_filename)
df = df.dropna().reset_index(drop=True)
```

```python
df.info()
```

```python
dfl = df[ [column for column in df.columns if "_NUM_" not in column] ]
```

```python
dfl.info()
```

```python
import sklearn.model_selection

train, test = sklearn.model_selection.train_test_split(
    dfl,
    test_size = 0.2,
    random_state= 0,
    shuffle = True,
    stratify=dfl["WINNER"]
)
```

```python
dfl["WINNER"].value_counts(), train["WINNER"].value_counts(), test["WINNER"].value_counts()
```

```python
train.reset_index(drop=True).to_csv("training_dataset.csv")
test.reset_index(drop=True).to_csv("testing_dataset.csv")
```
